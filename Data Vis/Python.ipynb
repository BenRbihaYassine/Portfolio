```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier Excel
df = pd.read_excel(r"C:\Users\Pc\Desktop\dataviz\New folder\Tunisie\manpartgvt.xlsx")

# D√©finir le style Seaborn
sns.set_style("whitegrid")

# Cr√©ation de la figure
plt.figure(figsize=(12, 6))

# Tracer un graphique √† barres pour les dossiers approuv√©s par r√©gion (GVT)
ax = sns.barplot(data=df, x="GVT", y="Dossiers approuv√©s ", palette="viridis")

# Ajouter des titres et des labels
plt.title("Dossiers Approuv√©s par R√©gion", fontsize=14)
plt.xlabel("R√©gion (GVT)", fontsize=12)
plt.ylabel("Nombre de Dossiers Approuv√©s", fontsize=12)
plt.xticks(rotation=45, ha="right")  # Rotation des labels de l'axe X si n√©cessaire

# Ajouter les valeurs sur chaque barre
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=10, color='black', 
                xytext=(0, 5),  # D√©calage du texte pour √©viter qu'il touche la barre
                textcoords='offset points')

# Afficher la visualisation
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier Excel
df = pd.read_excel(r"C:\Users\Pc\Desktop\dataviz\New folder\Tunisie\mansecteur.xlsx")

# S'assurer que 'Year' est trait√© comme une cha√Æne de caract√®res
df["Year"] = df["Year"].astype(str)

# Cr√©ation de l'histogramme
plt.figure(figsize=(10, 6))
bars = plt.bar(df["Year"], df["DOSSIERS IAA"], color='skyblue', edgecolor='black')

# Ajouter les valeurs au-dessus des barres
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, int(yval), ha='center', va='bottom', fontsize=9)

plt.title("R√©partition des Dossiers IAA par ann√©e (1996-2016)", fontsize=14)
plt.xlabel("Ann√©e")
plt.ylabel("Nombre de Dossiers IAA")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donn√©es
df = pd.read_csv(r"C:\Users\Pc\Desktop\data complet\data_COVID.csv") 
# Grouper par continent et calculer le total des cas
continent_cases = df.groupby('continentExp')['cases'].sum()

# Calculer les pourcentages
percentages = continent_cases / continent_cases.sum()

# Garder uniquement les continents avec au moins 1% des cas
filtered_cases = continent_cases[percentages >= 0.01]
# Tracer le pie chart avec l√©gende
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    filtered_cases,
    labels=None,  # Supprimer les labels internes
    autopct='%1.1f%%',
    startangle=140
)

# Ajouter la l√©gende avec couleurs
ax.legend(
    wedges,
    filtered_cases.index,
    title="Continents",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

# Tracer le graphique
plt.title("R√©partition des cas Covid par continent en 2020")
plt.axis('equal')
plt.show()
```
